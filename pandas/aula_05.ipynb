**üß† 1. Teoria simples**

No mundo real, dados v√™m de muitas fontes: planilhas do Excel, sistemas de vendas, bancos de dados, arquivos JSON, etc.
O pandas facilita a importa√ß√£o e exporta√ß√£o desses formatos com fun√ß√µes simples:

| Formato | Leitura           | Escrita         |
| ------- | ----------------- | --------------- |
| CSV     | `pd.read_csv()`   | `df.to_csv()`   |
| Excel   | `pd.read_excel()` | `df.to_excel()` |
| JSON    | `pd.read_json()`  | `df.to_json()`  |


Essas fun√ß√µes preservam a estrutura dos dados e permitem aplicar par√¢metros √∫teis como:

* `sep` (separador),

* `decimal` (formato de n√∫mero),

* `parse_dates` (converter datas automaticamente),

* `index_col` (definir √≠ndice).

**üíª 2. Demonstra√ß√£o pr√°tica**

```python
import pandas as pd

# Criando um DataFrame exemplo
vendas = pd.DataFrame({
    "Data": ["2025-01-01", "2025-01-02", "2025-01-03"],
    "Produto": ["Notebook", "Mouse", "Teclado"],
    "Pre√ßo": [3200.00, 89.90, 249.00],
    "Quantidade": [10, 50, 30]
})

# Salvando em CSV
vendas.to_csv("vendas.csv", index=False)

# Lendo o arquivo CSV
df_csv = pd.read_csv("vendas.csv")
print("Leitura do CSV:\n", df_csv)

# Salvando em Excel
vendas.to_excel("vendas.xlsx", index=False, sheet_name="Janeiro")

# Lendo planilha Excel
df_excel = pd.read_excel("vendas.xlsx", sheet_name="Janeiro")
print("\nLeitura do Excel:\n", df_excel)

# Salvando em JSON
vendas.to_json("vendas.json", orient="records", indent=4)

# Lendo JSON
df_json = pd.read_json("vendas.json")
print("\nLeitura do JSON:\n", df_json)
```

**‚öôÔ∏è Par√¢metros √∫teis em leitura**

```python
pd.read_csv("arquivo.csv", sep=";", decimal=",", parse_dates=["Data"], index_col="Data")
```

* `sep=";"` ‚Üí separador usado em CSVs do Excel brasileiro

* `decimal=","` ‚Üí v√≠rgula como separador decimal

* `parse_dates=["Data"]` ‚Üí converte strings em datas

* `index_col="Data"` ‚Üí define coluna de data como √≠ndice
